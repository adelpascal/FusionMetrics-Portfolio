```python
# Step 1: Define the beam and load details
span_length = 6  # Span length in meters
distributed_load = 500  # Load in N/m
point_load = 2000  # Point load in N
point_load_position = 3  # Position of the point load from the left in meters
# Step 2: Calculate reactions at supports
reaction_B = (distributed_load * span_length / 2) + (point_load * point_load_position) / span_length
reaction_A = (distributed_load * span_length) - reaction_B

print(f"Reaction at A: {reaction_A:.2f} N")
print(f"Reaction at B: {reaction_B:.2f} N")

```

```python
positions = np.linspace(0, span_length, 100)  # Positions along the beam
shear_forces = []
for x in positions:
    if x < point_load_position:
        shear_force = reaction_A - (distributed_load * x)
    else:
        shear_force = reaction_A - (distributed_load * x) - point_load
    shear_forces.append(shear_force)

print("Shear forces calculated.")
```

```python
# Step 4: Bending Moment Calculation
bending_moments = []
for x in positions:
    if x < point_load_position:
        moment = (reaction_A * x) - (distributed_load * x**2 / 2)
    else:
        moment = (reaction_A * x) - (distributed_load * x**2 / 2) - (point_load * (x - point_load_position))
    bending_moments.append(moment)

print("Bending moments calculated.")
```

```python
# Plot Shear Force Diagram
plt.figure(figsize=(10, 6))
plt.plot(positions, shear_forces, label="Shear Force", color="blue")
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")
plt.title("Shear Force Diagram")
plt.xlabel("Position (m)")
plt.ylabel("Shear Force (N)")
plt.legend()
plt.grid(True)
plt.show()

# Plot Bending Moment Diagram
plt.figure(figsize=(10, 6))
plt.plot(positions, bending_moments, label="Bending Moment", color="red")
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")
plt.title("Bending Moment Diagram")
plt.xlabel("Position (m)")
plt.ylabel("Bending Moment (Nm)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Beam and material properties
EI = 2e9  # Flexural rigidity (Nm)

# Deflection calculation using double integration
deflection = []
for x in positions:
    if x < point_load_position:
        deflection_value = (reaction_A * x**3 / (6 * EI)) - (distributed_load * x**4 / (24 * EI))
    else:
        deflection_value = ((reaction_A * x**3 / (6 * EI)) - (distributed_load * x**4 / (24 * EI))
                            - (point_load * (x - point_load_position)**3 / (6 * EI)))
    deflection.append(deflection_value)

# Plot Deflection Diagram
plt.figure(figsize=(10, 6))
plt.plot(positions, deflection, label="Deflection", color="green")
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")
plt.title("Deflection Diagram")
plt.xlabel("Position (m)")
plt.ylabel("Deflection (m)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Beam properties and load details
length = 7.5  # Beam length in meters
positions = [i * 0.1 for i in range(int(length / 0.1) + 1)]  # Discretize beam
distributed_load = 500  # Distributed load in N/m
point_load_position = 3.75  # Position of point load in meters
point_load = 2000  # Point load in Newtons

# Calculate Shear Force
shear_forces = []
for pos in positions:
    shear_force = -distributed_load * pos  # Contribution from distributed load
    if pos >= point_load_position:
        shear_force -= point_load  # Add point load contribution
    shear_forces.append(shear_force)

# Calculate Bending Moments by integrating Shear Forces
moments = []
current_moment = 0  # Initial bending moment
for i in range(len(positions) - 1):
    dx = positions[i + 1] - positions[i]  # Distance between points
    current_moment += shear_forces[i] * dx
    moments.append(current_moment)

# Add the last moment
moments.append(current_moment)

# Update Stress Calculation Code
I = 8e-6  # Moment of inertia (m^4)
y = 0.05  # Distance from neutral axis (m)

# Stress calculation
stresses = []
for moment in moments:
    stress = (moment * y) / I
    stresses.append(stress)

# Plot Stress Diagram
plt.figure(figsize=(10, 6))
plt.plot(positions, stresses, label="Normal Stress", color="purple")
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")
plt.title("Stress Distribution Along Beam")
plt.xlabel("Position (m)")
plt.ylabel("Stress (Pa)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Beam and material properties
E = 2e11  # Young's modulus in Pa
I = 8e-6  # Moment of inertia in m^4

# Deflection calculation
deflections = []
integration_constant = 0  # Assume zero for simplicity in this basic example
for pos in positions:
    # Deflection formula for distributed load
    deflection = (
        (-distributed_load * pos**4) / (24 * E * I)
        + (point_load * max(0, point_load_position - pos)**3) / (6 * E * I)
        + integration_constant
    )
    deflections.append(deflection)

# Plot Deflection Diagram
plt.figure(figsize=(10, 6))
plt.plot(positions, deflections, label="Deflection", color="blue")
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")
plt.title("Deflection Curve Along Beam")
plt.xlabel("Position (m)")
plt.ylabel("Deflection (m)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define beam parameters
L = 10  # Length of the beam in meters
E = 200e9  # Young's modulus in Pascals (steel)
I = 0.0001  # Moment of inertia in m^4
w = 500  # Uniform distributed load in N/m
P = 2000  # Point load in N
P_position = 5  # Position of the point load in meters
# Calculate reactions at supports (simplified for a simply supported beam)
R1 = (P * (L - P_position) + w * L * (L / 2)) / L
R2 = (P * P_position + w * L * (L / 2)) / L

print(f"Reaction at Support 1 (R1): {R1:.2f} N")
print(f"Reaction at Support 2 (R2): {R2:.2f} N")
```

```python
# Discretize the beam
x = np.linspace(0, L, 100)  # Beam length divided into 100 segments

# Shear force distribution
V = np.piecewise(x,
                 [x < P_position, x >= P_position],
                 [lambda x: R1 - w * x, lambda x: R1 - w * x - P])

print("Shear Force Distribution Calculated.")
```

```python
# Bending moment distribution
M = np.piecewise(x,
                 [x < P_position, x >= P_position],
                 [lambda x: R1 * x - (w * x ** 2) / 2,
                  lambda x: R1 * x - (w * x ** 2) / 2 - P * (x - P_position)])

print("Bending Moment Distribution Calculated.")
```

```python
# Plot Shear Force
plt.figure(figsize=(10, 6))
plt.plot(x, V, label="Shear Force", color='blue')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.title("Shear Force Distribution")
plt.xlabel("Position along the beam (m)")
plt.ylabel("Shear Force (N)")
plt.grid(True)
plt.legend()
plt.show()

# Plot Bending Moment
plt.figure(figsize=(10, 6))
plt.plot(x, M, label="Bending Moment", color='red')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.title("Bending Moment Distribution")
plt.xlabel("Position along the beam (m)")
plt.ylabel("Bending Moment (Nm)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Calculate deflection using double integration
dx = L / 100  # Step size for numerical integration

# Initialize deflection and slope arrays
slope = np.zeros(len(x))
deflection = np.zeros(len(x))

# Numerical integration for slope and deflection
for i in range(1, len(x)):
    slope[i] = slope[i - 1] + (M[i - 1] / (E * I)) * dx
    deflection[i] = deflection[i - 1] + slope[i - 1] * dx

# Convert deflection to mm for better visualization
deflection_mm = deflection * 1000

print("Deflection Calculation Completed.")
```

```python
# Plot Deflection
plt.figure(figsize=(10, 6))
plt.plot(x, deflection_mm, label="Deflection", color='purple')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.title("Deflection of the Beam")
plt.xlabel("Position along the beam (m)")
plt.ylabel("Deflection (mm)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Beam section dimensions
h = 0.3  # Height of the beam in meters
y_top = h / 2  # Distance from neutral axis to top fiber
y_bottom = -h / 2  # Distance from neutral axis to bottom fiber

# Normal stress calculation
stress_top = M * y_top / I
stress_bottom = M * y_bottom / I

print("Stress Calculation Completed.")
```

```python
# Plot Stress Distribution
plt.figure(figsize=(10, 6))
plt.plot(x, stress_top / 1e6, label="Top Fiber Stress", color='orange')
plt.plot(x, stress_bottom / 1e6, label="Bottom Fiber Stress", color='green')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.title("Normal Stress Distribution Along the Beam")
plt.xlabel("Position along the beam (m)")
plt.ylabel("Stress (MPa)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Summary of results
max_deflection = np.max(np.abs(deflection_mm))
max_stress_top = np.max(np.abs(stress_top)) / 1e6  # Convert to MPa
max_stress_bottom = np.max(np.abs(stress_bottom)) / 1e6  # Convert to MPa

print(f"Maximum Deflection: {max_deflection:.2f} mm")
print(f"Maximum Stress (Top Fiber): {max_stress_top:.2f} MPa")
print(f"Maximum Stress (Bottom Fiber): {max_stress_bottom:.2f} MPa")
```

```python
# Step 1: Foundation and Soil Properties
foundation_length = 2  # Length of the foundation in meters
foundation_width = 2   # Width of the foundation in meters
foundation_area = foundation_length * foundation_width  # Area in m

total_load = R1 + R2  # Reaction forces calculated earlier (in Newtons)
safe_bearing_capacity = 150e3  # Safe bearing capacity of soil in N/m (example value)

# Step 2: Calculate Bearing Pressure
bearing_pressure = total_load / foundation_area

# Step 3: Compare with Safe Bearing Capacity
if bearing_pressure > safe_bearing_capacity:
    print("Foundation dimensions are inadequate. Consider increasing the area.")
else:
    print("Foundation is stable under the given load.")

# Display Results
print(f"Total Load on Foundation: {total_load:.2f} N")
print(f"Foundation Area: {foundation_area:.2f} m")
print(f"Bearing Pressure: {bearing_pressure:.2f} N/m")
print(f"Safe Bearing Capacity: {safe_bearing_capacity:.2f} N/m")
```

```python
# Step 1: Define Material and Labor Costs
concrete_cost_per_m3 = 750_000  # IDR/m
steel_cost_per_ton = 10_000_000  # IDR/ton
labor_cost_per_day = 500_000  # IDR/person/day
construction_days = 30  # Duration of construction

# Step 2: Material Quantities
# Concrete volume for columns and beams
column_volume = 0.3 * 0.3 * 4 * 4  # 4 columns, 0.3m x 0.3m x 4m each
beam_volume = 0.3 * 0.2 * 5 * 4  # 4 beams, 0.3m x 0.2m x 5m each
total_concrete_volume = column_volume + beam_volume  # Total concrete volume in m

# Steel weight for reinforcement
steel_density = 7850  # kg/m
steel_volume = total_concrete_volume * 0.02  # Assume 2% of concrete volume is steel
total_steel_weight = steel_volume * steel_density / 1000  # Convert to tons

# Step 3: Cost Calculations
total_concrete_cost = total_concrete_volume * concrete_cost_per_m3
total_steel_cost = total_steel_weight * steel_cost_per_ton
total_labor_cost = labor_cost_per_day * construction_days * 5  # Assuming 5 workers

# Total cost
total_cost = total_concrete_cost + total_steel_cost + total_labor_cost

# Display Results
print(f"Concrete Volume: {total_concrete_volume:.2f} m")
print(f"Steel Weight: {total_steel_weight:.2f} tons")
print(f"Concrete Cost: {total_concrete_cost:.2f} IDR")
print(f"Steel Cost: {total_steel_cost:.2f} IDR")
print(f"Labor Cost: {total_labor_cost:.2f} IDR")
print(f"Total Construction Cost: {total_cost:.2f} IDR")
```

```python
# Number of nodes and elements
nodes = 4
elements = 3

# Length of each element (in meters)
length = 5.0

# Stiffness of each element (in N/m)
stiffness = 2000

# Global stiffness matrix initialization
K = np.zeros((nodes, nodes))

# Assemble the stiffness matrix
for i in range(elements):
    K[i, i] += stiffness
    K[i, i + 1] -= stiffness
    K[i + 1, i] -= stiffness
    K[i + 1, i + 1] += stiffness

# Apply boundary conditions (fixed at node 0)
K_reduced = K[1:, 1:]  # Remove first row and column for fixed node

# Apply external forces
forces = np.zeros(nodes)
forces[2] = 500  # Apply force of 500 N at node 2
forces_reduced = forces[1:]  # Adjust for reduced matrix

# Solve for displacements
displacements = np.linalg.solve(K_reduced, forces_reduced)

# Calculate reactions
reactions = np.dot(K, np.concatenate(([0], displacements)))

print("Displacements at nodes:", displacements)
print("Reactions at supports:", reactions[0])
```

```python
# Mass matrix (diagonal for simplicity)
mass = np.array([1000, 1000, 1000])  # Mass of each node in kg

# Solve the eigenvalue problem (Kx = Mx)
eigvals, eigvecs = np.linalg.eig(np.linalg.inv(np.diag(mass)) @ K_reduced)

# Calculate natural frequencies
frequencies = np.sqrt(eigvals) / (2 * np.pi)

print("Natural Frequencies (Hz):", frequencies)
print("Mode Shapes (Normalized):", eigvecs)
```

```python
# Combine wind forces with other loads
forces_combined = forces_reduced + np.array([horizontal_wind_force, vertical_wind_force, 0])

# Solve for displacements under combined loads
displacements_combined = np.linalg.solve(K_reduced, forces_combined)

print("Displacements under combined loads:", displacements_combined)
```

```python
# Nonlinear stiffness (assume stiffness reduces by 10% per 1m of displacement)
def nonlinear_stiffness(displacement, initial_stiffness):
    return max(initial_stiffness * (1 - 0.1 * displacement), 0)

# Iterative solver for nonlinear stiffness
displacements_nonlinear = np.zeros(len(forces_reduced))
tolerance = 1e-5
max_iterations = 100

for iteration in range(max_iterations):
    stiffness_nonlinear = np.array([nonlinear_stiffness(d, stiffness) for d in displacements_nonlinear])
    K_nonlinear = np.diag(stiffness_nonlinear)
    displacements_new = np.linalg.solve(K_nonlinear, forces_reduced)
    
    if np.allclose(displacements_nonlinear, displacements_new, atol=tolerance):
        break
    displacements_nonlinear = displacements_new

print("Displacements under nonlinear loads:", displacements_nonlinear)
```
